sin(а*x)=0
при условии: 0.01<a<0.02, 100<х<500.
Т.е. надо найти решение х как функцию параметра а - построить график x=x(а).

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
a = np.linspace(0.01, 0.02, 1000)
x = np.linspace(100, 500, 1000)
```

```python
plt.plot(a, np.pi/a, label='x=np.pi/a')
plt.plot(a, np.pi*2/a, label='x=np.pi*2/a')
plt.plot(a, np.pi*3/a, label='x=np.pi*3/a')

plt.xlabel("a")
plt.ylabel("x")
plt.ylim(100,500)

plt.legend(loc='upper left')

plt.show()
```
