 Постройте на одном графике две кривые y(x) для функции двух переменных y(k,x)=cos(k∙x), взяв для одной кривой значение k=1, а для другой – любое другое k, не равное 1.

```python
import numpy as np
import matplotlib.pyplot as plt
from pylab import rcParams

%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
x = np.linspace(-5, 5, 101)
y1 = np.cos(x)
y2 = np.cos(4*x)
```

```python
rcParams['figure.figsize'] = 8, 5
```

```python
plt.plot(x, y1, label='y(1,x)=cos(x)')
plt.plot(x, y2, label='y(4,x)=cos(4x)')

plt.xlabel("Переменная x")
plt.ylabel("Функция y(k,x)")

plt.legend(loc='upper right')

plt.show()
```
