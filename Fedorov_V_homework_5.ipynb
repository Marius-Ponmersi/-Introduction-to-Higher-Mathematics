4. (не обязательно, но желательно) Из урока по комбинаторике повторите расчеты, сгенерировав возможные варианты перестановок для других значений n и k

```python
import itertools
```

Размещения

```python
permutation = []
for p in itertools.permutations("123456", 3):
    permutation.append(p)
    print(' '.join(str(x) for x in p))
print(f'len(permutation) = {len(permutation)}')
```

Сочетания

```python
combination = []
for p in itertools.combinations("123456", 3):
    combination.append(p)
    print(''.join(p))
print(f'len(combination) = {len(combination)}')
```

Перестановки (n=k)

```python
permutation = []
for p in itertools.permutations("123456", 6):
    permutation.append(p)
    print(' '.join(str(x) for x in p))
print(f'len(permutation) = {len(permutation)}')
```

5. (не обязательно) Дополните код расчетом коэффициента корреляции x и y по формуле

```python
n = 100
r = 0.7
x = np.random.rand(n)
y = r*x + (1 - r)*np.random.rand(n)
plt.plot(x, y, 'o')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)

a = (np.sum(x)*np.sum(y) - n*np.sum(x*y))/(np.sum(x)*np.sum(x) - n*np.sum(x*x))
b = (np.sum(y) - a*np.sum(x))/n

A = np.vstack([x, np.ones(len(x))]).T
a1, b1 = np.linalg.lstsq(A, y)[0]

c = np.corrcoef(x, y)

c_cor = np.sum((x - np.mean(x))*(y - np.mean(y))) / math.sqrt(np.sum((x - np.mean(x))**2)*np.sum((y - np.mean(y))**2))

print(f'a = {a}, b = {b}')
print(f'a = {a1}, b = {b1}\n')

print(f'Коэффициент корреляции: {c} \n')
print(f'Коэффициент корреляции: {c_cor}')

plt.plot([0, 1], [b, a + b])
plt.show()
```
